# Linear Regression:
Linear Regression is a supervised learning algorithm used for predicting a continuous target variable based on one or more input features by modeling the linear relationship between them
<br>`Equation: y = mx + c`
<br>
This is the equation of a straight line â€” and it's the core idea behind Linear Regression, which is one of the simplest and most widely used algorithms in Machine Learning.

* y	The predicted value (output)
* x	The input feature (independent variable)
* m	The slope of the line (also called the weight or coefficient)
* c	The y-intercept (also called the bias)

```python
import pandas as pd
data = {
    'area': [750, 800, 850, 900, 950, 1000, 1050, 1100, 1150, 1200],
    'price': [150000, 160000, 170000, 180000, 190000, 200000, 210000, 220000, 230000, 240000]
}

df = pd.DataFrame(data)
print(df)

# ...existing
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression   
```

```python
%matplotlib inline
plt.scatter(df.area, df.price,color='green', marker='*')
plt.xlabel('Area (sq ft)')  
plt.ylabel('Price ($)')
```

```python
reg_model = LinearRegression()
reg_model.fit(df[['area']], df['price'])
```

```python
reg_model.predict([[830]])
```
